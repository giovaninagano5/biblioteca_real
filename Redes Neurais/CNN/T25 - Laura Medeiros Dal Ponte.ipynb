## Revisando Redes Neurais Artificiais

O conceito de redes neurais artificais toma inspiração direta da neurologia. Nesse modelo matemático, atributos são análogos a neurônios que passam informações adiante em sinapses artificais e resultam em um dado ao final. A principal vantagem dessas redes é o fato de que quanto maior for o fluxo de dados atribuído a ela, melhor será sua performance, uma vez que ela aprende a detectar padrões automaticamente. Por isso, elas são amplamente usadas para problemas de classificação, previsão, regressão, agrupamento, entre outros.

Uma rede neural funciona, basicamente, da seguinte forma:

1) Primeiro, ela recebe os valores de entrada
2) Ela multiplica esses valores por um peso (W)
3) Em seguida, ela soma esses valores junto com um viés (b)
4) A soma resultante é enviada para uma função de ativação que introduz complexidade ao modelo
5) Por fim, envia o resultado para a próxima camada

Durante o treinamento, ela ajusta os pesos e vieses para reduzir o erro em relação a entrada e saída prevista e melhorar o desempenho.

## Redes Neurais Convolucionais (CNN)

A rede neural do tipo **convolucional** (CNN) é uma arquitetura projetada para processamento de data com topologia "de grade", como imagens, capturando padrões espaciais por meio de operações convolucionais, tornando-as particularmente interessante quando temos problemas relacionados a reconhecimento de objeto, detecção, classificação de imagem e segmentação. <a href="#ref-4">[2]</a><a href="#ref-4">[3]</a>. Entre algumas de suas aplicações no cenário científico real, ganha destaque na área da saúde (para diagnóticos, detecção de tumores, análise de MRI e de células), da microbiologia (no reconhecimento de patógenos, vírus e bactérias) e na ciência computacional como um todo (reconhecimento de gestos, expressões e rostos). 

Apesar de serem mais popularmente conhecidas por trabalharem com imagens, as CNN's são incrivelmente versáteis e podem também serem aplicadas em análises temporais, análises de sequências genéticas, reconhecimento de fala e processamento de linguagem natural. Essas redes são classificadas como **CNN's de 1 dimensão (1D)**, enquanto aquelas que trabalham com imagens são classificadas como **CNN's de 2 dimensões (2D)**. Existem ainda as **CNN's de 3 dimensões (3D)**, que são utilizadas para lidar com dados volumétricos ou sequências temporais de imagens e vídeos <a href="#ref-4">[4]</a>. **Esse notebook em particular irá focar em redes neurais 2D**, mas mencionaremos as demais quando conveniente.

As CNN's são particularmente poderosas em extrair características de dados oferecidas. Elas conseguem automaticamente identificar padrões e classificar dados independente de variações em seu posicionamento, escala e orientação. Por tabela, elas também são ótimas para trabalhar com um grande volume de dados, mas possuem alto custo computacional dependendo do seu objetivo com ela, sendo quase sempre necessário o uso de GPU's para um processamento adequado. 

A estrutura básica de uma CNN tem a seguinte anatomia:

- **Camadas convolucionais**: onde ocorrem operações convolucionais sobre os dados de entrada. É aqui que ocorre o processo de extração de features/características das imagens.
- **Função de ativação**: introduz não linearidade e permite que a rede aprenda relações complexas entre os dados.
- **Camadas de *pooling***: agrupa os dados de saída das camadas convolucionais mais relevantes e informacionais.
- **Camadas integralmente conectadas**: etapa onde são feitas as previsões, com dados de entrada correspondentes a uma matriz achatada unidimensional.

As camadas pertencentes a uma CNN são chamadas de **camadas convolucionais**. Além disso, no córtex frontal, os neurônios estão conectados a uma região local chamada **campos receptivos**. ### Mas afinal, o que é convolução?

Matematicamente falando, **operadores convolucionais** são aqueles que, a partir de duas funções dadas, resultam em uma terceira. Essa nova função mostra o quanto uma função “se encaixa” na outra à medida que uma delas é deslocada no tempo ou no espaço, onde em cada posição desse deslocamento, calcula-se a soma dos produtos dos valores das duas funções onde elas se sobrepõem. As convoluções podem ser tanto contínuas (definidas em um domínio espaço-tempo contínuo) ou, para o caso aqui, discretas. As convoluções discretas trabalham com sinais definidos em instantes discretos que aqui chamamos de amostras. Uma operação convolucional discreta é expressa como <a href="#ref-4">[5]</a>:

$$
(f * g)(k) = h(k) = \sum_{j=0}^{k} f(j)\cdot g(k - j) 
$$

Em termos simples: duas funções podem ser combinadas para gerar uma nova função - ou melhor, dois sinais podem resultar em um terceiro novo sinal. A medida que novos sinais vão sendo gerados, mais informativos eles se tornam. Cada tipo rede convolucional (1D, 2D e 3D) calcula convolução de acordo com sua estrutura dimensional: vetores para 1D, matrizes para 2D e volume para 3D. Porém, para todas as operações convolucionais, um **kernel** é utilizado para deslizar sobre os dados de entrada, multiplicando elemento a elemento e gerando resultados.

### Kernels 

Invés da CNN trabalhar com a imagem inteira de uma vez só, a rede neural trabalha com uma pequena fração da mesma, chamada **kernel** ou **filtro**. É aqui que acontecem as operações de convolução, onde o kernel irá "deslizar" sobre o dado recebido pela rede. Uma rede neural convolucional é composta por vários desses filtros, cada um observando a imagem por sua própria "janela". A área máxima de imagem que uma rede convolucional consegue computar é o que conhecemos como o **campo receptivo da CNN.** Na primeira camada, cada neurônio vê apenas a área correspondente a esse campo. A medida esse dado atravessa camadas subsequentes, esse campo tende a aumentar progressivamente, permitindo a visualização de pelo menos uma boa parte da imagem original. <a href="#ref-4">[6]</a> Depois que o kernel é aplicado sobre a imagem toda, temos o que chamamos de um **feature map**, que contêm o resultado das convoluções resultantes.

### Camadas de *pooling*

As camadas de *pooling* entram logo após as operações convolucionais. Elas servem para "resumir" os feature maps que foram gerados pelos kernels, buscando focar em elementos mais significantes e informativos da imagem. Algumas das técnicas conhecidas de pooling são <a href="#ref-4">[3]</a>: 

- `MaxPooling`: a camada retorna os maiores valores dentro da janela a qual lhe é atribuída.
- `PlusPooling`: a camada retorna a soma dos valores pertencentes a janela a qual lhe é atribuída.
- `AveragePooling`: a camada faz a média dos valores pertecentes a janela a qual lhe é atribuída.

Há algumas transformações que podemos aplicar para nossas imagens, as quais estão descritas abaixo. Além da prática clássica de converter as imagens para um dado com qual o Pytorch consiga trabalhar (tensores) e normalizar os dados, podemos também optar por unificar o número de canais ao converter as imagens para uma escala de cinza. Além disso, dentro do conjunto teste temos algumas transformações que introduzem maior variedade aos dados e permitem que o modelo faça previsões mais generalizadas. Essa prática em particular é conhecida como **data augmentation**.

- `transform.Resize()`: redimensionamento das imagens (pixels) - isso pode ser tanto para garantir que todas as imagens tem o mesmo tamanho quanto para diminuir o tamanho dos seus dados (as vezes pode ser útil para exploração).
  
- `transform.Grayscale()`: converte as imagens para escala de cinza, reduzindo o número de canais para um único canal. Isso pode reduzir a complexidade das imagens, o que talvez seja interessante para um aprendizado mais estável, embora limite o modelo de capturar relações entre cores.
  
- `transform.RandomRotation()`: rotaciona a imagem aleatoriamente em um ângulo pré definido. Ângulos menores (entre 10 e 25 graus) são recomendados, uma vez que ajudam a aumentar a generalização sem distorcer excessivamente os dados.
  
- `transform.RandomHorizontalFlip()`: espelha sua imagem na horizontal aleatóriamente.
  
- `transforms.ToTensor()`: transforma a imagem em tensores compatíveis para Pytorch, normalizando os valores dos pixels para intervalos entre 0 e 1.
  
- `transforms.Normalize()`: aplica normalização nos tensores utilizando médias e desvios pré definidos. 

```python
train_transforms = transforms.Compose([
    transforms.Resize((28, 28)), 
    transforms.Grayscale(num_output_channels=1),
    transforms.RandomRotation(15), 
    transforms.RandomHorizontalFlip(), 
    transforms.ToTensor(), 
    transforms.Normalize([0.5], [0.5]) 
    
])

test_transforms = transforms.Compose([
    transforms.Resize((28, 28)), 
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(), 
    transforms.Normalize([0.5], [0.5])

])
```

### Criando DataLoaders

Dataloaders são cruciais para uma boa parte das redes neurais, introduzindo os **batches**, pequenos conjuntos de amostras sobre qual a rede trabalha. Em batches de treino, a técnica de **shuffle** é recomendada, uma vez que ela embaralha os dados e ajuda a rede a generalizar melhor.

Você pode perceber que as etapas de criação de dataloaders e treinar o modelo são muito parecidas com as de uma MLP tradicional, o que é verdade: a principal diferença está em como montamos a arquitetura do modelo.

```python
def Loaders(train_dataset, val_dataset, batch_size, num_workers=0):
    
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size, 
        shuffle=True, 
        num_workers=num_workers)
    
    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size*2, 
        shuffle=False, 
        num_workers=num_workers)
    
    return train_loader, val_loader
```

### Criando um modelo de CNN

Essa seção foi baseada na playlist didática produzida pelo canal *John Elder*. Deixo o link para a mesma na referência </a><a href="#ref-4">[7]</a>.

Como mencionado previamente, uma arquitetura padrão de CNN segue o seguinte fluxo: **Camadas convolucionais** → **Camadas de pooling** → **Camadas integradas** → **Previsão**

Em `self.features`, definimos a arquitetura geral das nossas camadas convolucionais e como será feita a extração de features. A primeira camada convolucional da CNN irá receber os canais da imagem de entrada. Como estamos trabalhando com escala cinza, temos somente um canal de entrada (c_in=1). 

Definimos então os canais de saída que são gerados e posteriormente alimentados para a próxima camada. **O número de canais de saída corresponde ao número de kernels que utilizamos na imagem**, gerando um canal por kernel. O tamanho dos kernels, que define quanta informação eles devem extrair da imagem (2x2, 3x3, 4x4, etc), também é um hiperparâmetro otimizável. 

**O parâmetro `padding` ajuda a controlar o tamanho das saídas e preserva informação sobre as bordas.** Essa técnica consistem em preencher os arredores das bordas de uma imagem com valores extras (valores pequenos ou nulos) antes da convolução ser aplicada. Sem isso, as imagens encolheriam até que alcançassem um tamanho onde não seria possível extrair nenhuma informação sobre elas. O número que atribuímos a ele corresponde a espessura da moldura que será aplicada a imagem (que no caso, está definida para 1 pixel).

Para ambas as camadas, aplicamos a função de ativação `ReLU()`, que define o intervalo $A(x) = max(0, x)$. Logo em seguida definimos uma **camada de pooling** que captura as informações mais relevantes dos outputs gerados pelas camadas convolucionais. A técnica de `MaxPooling` consiste em pegar os maiores valores de uma região da imagem. O parâmetro `stride` define o tamanho da janela (extensão da região a ser observada) na qual ela irá selecionar valores. No nosso caso, consideramos uma janela de 2x2 pixels de extensão.

Na última camada, aplicamos a técnica de **Dropout**, que consiste em aleatoriamente desligar alguns neurônios durante a fase de treino. A implementação é idêntica a como fazemos em uma MLP. Veremos adiante que outros reguladores e técnicas, como reguladores por peso e **Early stopping** também são aplicáveis para CNN's!

Já no bloco `self.head`, definimos como ocorrem as decisões finais (classificações das imagens). Após achatamento das matrizes 2D (flattening), o tensor é transformado em um vetor unidimensional que percorre a camada de neurônios final. A dimensionalidade desse vetor é definida como 37 x 7 x 7 (valores puramente idealizados para esse exemplo). Ele é processado por uma camada de 128 neurônios e, finalmente, passa pela camada de saída, gerando previsões por intermédio de uma função `SoftMax`. Essa é uma função de ativação converte os logits da rede para valores entre 0 e 1, interpretando as saídas como probabilidades de cada classe. 

```python
class CNN(nn.Module):
    def __init__(self, c_in):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(
            c_in, 
            16, 
            kernel_size=3, 
            padding=1
            ),
            
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(
            16, 
            32, 
            kernel_size=3, 
            padding=1
            ),
            
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Dropout2d(0.2),
        )

        self.head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 7 * 7, 128),
            nn.ReLU(),
            nn.Linear(128, 8)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.head(x)
        return x
```

### Integrando o modelo no Lightning

O `Lightning` nos oferece alguns recursos de regularização que podem ser interessantes para a nossa rede, além de nos poupar algumas linhas de código. Nesse caso, usamos a regularização por `weight_decay` (adiciona penalidade para pesos grandes) e o método de `EarlyStopping`, que permite que o treinamento por épocas seja interrompido assim que o modelo para de apresentar melhoras. Além disso, podemos instanciar **callbacks** e **métricas de acurácia** que nos fornecem informação quando a performance do modelo ao longo da data. 

Para nossa função de perda, usamos a `CrossEntropyLoss()`. Essa é a função de perda mais utilizada para problemas de classificação binária e multiclasse. Ela funciona comparando as pontuações feitas pelo modelo (logits) com a classe correta, aumentando a perda a medida que a densidade de probabilidade atribuída a classe correta (dado a previsão do modelo) é baixa </a><a href="#ref-4">[8]</a>. Para a acurácia, utilizamos a métrica de `MulticlassAccuracy` disponibilizada pelo `Lightning`. Escolhemos o otimizador `Adam` para atualizar os pesos após o cálculo do gradiente. 

A escolha de valores para learning rate e weight_decay são apenas exemplos - é altamente recomendado otimizá-los!

```python
class LitWrap(L.LightningModule):
    def __init__(
        self, modelo, lr = 1e-3, optmizer_name = "adam", weight_decay = 1e-4
    ):
        super().__init__()

        self.modelo = modelo
        self.lr = lr
        self.loss_fn = nn.CrossEntropyLoss()
        self.optmizer_name = optmizer_name
        self.weight_decay = weight_decay
        self.val_acc = MulticlassAccuracy(num_classes = 8)


    def forward(self, x):
        return self.modelo(x)

    def configure_optimizers(self):
        
         optimizer = torch.optim.Adam(
         self.parameters(),
         lr=self.lr,
         weight_decay=self.weight_decay
         )
        
         return optimizer
     
    def training_step(self, batch, batch_idx):
        x, y = batch
        y = y.squeeze().long()
        logits = self(x)
        loss = self.loss_fn(logits, y)
        
        self.log("train_loss", loss, on_epoch=True, prog_bar=False)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y = y.squeeze().long()
        logits = self(x)
        loss = self.loss_fn(logits, y)

        
        preds = torch.argmax(logits, dim=1)
        acc = self.val_acc(preds, y)


        self.log("val_loss", loss, on_epoch=True, prog_bar=True)
        self.log("val_acc", acc, on_epoch=True, prog_bar=True)
        return loss
```

Formalizamos um `Callback` que irá armazenar as perdas de treino e validação. 

```python
class Historico(Callback):
    
    def __init__(self):
    
        self.train_losses = []
        self.val_losses = []

    def on_train_epoch_end(self, trainer, pl_module):
        tl = trainer.callback_metrics.get("train_loss")
        if tl is not None:
            self.train_losses.append(float(tl.detach().cpu()))

    def on_validation_epoch_end(self, trainer, pl_module):
        vl = trainer.callback_metrics.get("val_loss")
        if vl is not None:
            self.val_losses.append(float(vl.detach().cpu()))
```

Finalmente, chamamos nossa rede e realizamos o treinamento final com `Lightning.Trainer`. Aqui também implementamos o `EarlyStopping`.

```python
train_loader, val_loader = Loaders(
    train_dataset_t,
    val_dataset_t,
    batch_size=64
)

cnn = CNN(c_in=1)


lit_model = LitWrap(
    modelo=cnn,
    lr=1e-3
)

hist_loss = Historico()

early_stop = EarlyStopping(
    monitor="val_loss",  
    patience=5,          
    mode="min",           
)

trainer = L.Trainer(
    max_epochs=100,
    accelerator="auto",
    devices="auto",
    callbacks=[hist_loss, early_stop],
    log_every_n_steps=10
)

trainer.fit(
    lit_model,
    train_dataloaders=train_loader,
    val_dataloaders=val_loader
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(hist_loss.train_losses, label="Treino")
plt.plot(hist_loss.val_losses, label="Validação")
plt.xlabel("Época")
plt.ylabel("Loss")
plt.title("Loss de Treino vs Validação")
plt.legend()
plt.grid(True)
plt.show()
```

Com o callback, podemos também recuperar a acurácia calculada pela nossa métrica:

```python
val_acc = trainer.callback_metrics["val_acc"].item()
print(f"Acurácia de validação: {val_acc:.4f}")
```

Redes Neurais Convolucionais são muito eficazes para tarefas de extração de features e processamento de um grande volume de dados, principalmente para problemas de visualização computacional para detecção e classificação de objetos.
