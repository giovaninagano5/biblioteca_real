## Introdução

A partir do reconhecimento da importância das redes neuris, tem se apresentado cada vez mais relevante aplicações em que os dados são gerados a partir de domínios não euclidianos e são representados como grafos com relações complexas e interdependência entre objetos [1]. Voltando brevemente ao que se entende como grafo, é possível interpretá-lo como uma forma de expressar um conjunto de objetos e as conexões entre eles [2]. Nesse sentido quando aplicados a redes neurais existem três tipos gerais de tarefas de predição em grafos: em nível de grafo, em nível de nó e em nível de aresta [2].

Elaborando-se uma breve linha temporal, Sperduti et al. (1997) [3] aplicaram pela primeira vez redes neurais a grafos acíclicos direcionais, o que motivou os primeiros estudos sobre Graph Neural
Networks (GNNs). A noção de redes neurais gráficas que foi inicialmente desenvolvida em Gori et al. (2005) [4] e posteriormente elaborada em Scarselli et al.(2009) [5] e Gallicchio et al. (2010) [6]. Esses primeiros estudos se enquadram na categoria de redes neurais gráficas recorrentes (RecGNNs) [1].

As práticas atuais evoluiram consideravelmente com isso existe variações de GNN, como redes convolucionais de grafos (GCNs), redes de atenção de grafos (GATs) e GraphSAGE, quem têm demonstrado desempenho inovador em várias tarefas de aprendizado profundo nos últimos anos [7]. Existem muitas escolhas de projeto que devem ser feitas ao usar essas arquiteturas, o sucesso das GNNs nos últimos anos cria uma grande oportunidade para uma ampla gama de novos problemas [2].

Com um foco mair nas GCNs, elas podem ser consideradas redes neurais convolucionais que foram expandidas para lidar com dados estruturados em grafos [7]. Nesse sentido, as operações convolucionais são adaptadas para lidar com estruturas irregulares e não semelhantes a grades de grafos, essas operações normalmente envolvem a agregação de informações de nós vizinhos para atualizar as características de um nó central [7]. Esses modelos escalam linearmente no número de edges do grafo, tornando-os adequados para grandes conjuntos de dados. 

Pra construção de um estudo mais aprofundado de GCN, o trabalho a seguir foi elaborado com o objetivo de aplicar uma rede neural convolucionail de grafos a dados de pangenômica de da bactéria *Helicobacter pylori*. Afim de que contase tanto com dados biológicos reais, quanto com o foco na parte computacional do modelo, sendo classificado os dados genômicos em três categorias (raro, moderado, comum) com base em seu padrão de presença/ausência nos genomas e, crucialmente, nas relações de co-ocorrência entre eles (se dois genes tendem a aparecer juntos nos mesmos genomas). A ideia central é que genes co-ocorrentes podem estar envolvidos em processos biológicos similares ou em módulos funcionais, ajudando assim a identificar uma padrão para construção do pangenoma, e a GCN é um possibilidade para capturar essa informação estrutural [9, 10]. 

## Glossário

Para melhor entendimento do material apresentado nesse notebook é interesante reforçar pelo glossário que associa as definições biológicas e computacionais.

**Grafo (Graph):** Uma representação que mostra como diferentes tipos de informação estão relacionadas, composto por nós (genes) e arestas (coocorrências entre genes) [11].

**Nó (Node):** Elemento individual do grafo que representa uma entidade específica. Neste estudo, cada nó corresponde a uma família gênica presente no pangenoma de *Helicobacter pylori* [1, 11].

**Características dos Nós (Node Features):** Cada nó tem uma "identidade" para um gene, sendo simplesmente um vetor de em quais genomas ele aparece [9].

**Aresta (Edge):** Conexão entre dois nós do grafo. As arestas foram construídas com base na coocorrência gênica, isto é, genes presentes simultaneamente em um número mínimo de genomas [1,11].

**Matriz de Presença e Ausência:** Representação binária em que linhas correspondem a genes e colunas a genomas. Valores iguais a 1 indicam presença do gene, enquanto valores iguais a 0 indicam ausência [9,15].

**Convolução (Convolution):** A GCN olha para um nó, pega as características de todos os seus vizinhos, faz uma média dessas características, e combina com as características do próprio nó. Isso gera uma nova representação para o nó [11].

**Camadas (Layers):** O processo de convolução reptidadas vezes gera camadas, após duas camadas, um nó já contém informações dos nós a 2 passos de distância [11]. 

**Pangenoma** Conjunto de genes presentes em todas as linhagens sequenciadas de uma espécie, associados pelo genoma de linhagem mais próximo por padrão genético [12]. 

**Limiar (Threshold):** Valor mínimo utilizado para definir se uma relação entre dois genes será considerada significativa na construção do grafo. Neste trabalho, o limiar representa o número mínimo de genomas em que dois genes precisam coexistir para que uma aresta seja criada entre eles [11].

**Coocorrência Gênica:** Medida que indica quantas vezes dois genes aparecem simultaneamente nos mesmos genomas [11].

**Densidade do Grafo (Graph Density):** Medida que representa o quão conectado um grafo está em relação ao número máximo possível de conexões. Uma alta densidade pode indicar forte compartilhamento gênico entre linhagens bacterianas, enquanto densidades menores sugerem maior diversidade genética [1, 7].

**Embedding:** Representação vetorial aprendida pela rede neural para cada nó do grafo. Genes com funções ou padrões evolutivos semelhantes tendem a apresentar embeddings próximos no espaço vetorial aprendido pela GCN [1, 7].

**Dropout:** Técnica de regularização utilizada para reduzir overfitting, desativando aleatoriamente parte dos neurônios durante o treinamento [1].

**Grau do Nó (Node Degree):** Número de conexões associadas a um nó [1]. Genes com alto grau tendem a compartilhar padrões de presença com muitos outros genes, podendo representar genes centrais ou altamente conservados no pangenoma [11].

### Construir a matriz de co-ocorrência

Nessa etapa foi importante (i) calcular uma medida de coexpressão entre cada par de genes e (ii) selecionando um limiar de significância para remover relações espúrias entre os genes [11]. A abordagem adotada fundamentou-se no princípio de que genes que co-ocorrem frequentemente nos mesmos genomas tendem a estar envolvidos em processos biológicos similares ou a fazer parte de módulos funcionais conservados [10].

O cálculo da matriz seguiu o procedimento descrito em [11], onde a matriz de presença/ausência (dimensões n_genes × n_genomas) é multiplicada por sua transposta, resultando em uma matriz quadrada n_genes × n_genes cujos elementos representam o número de genomas em que cada par de genes aparece simultaneamente. Para construir a matriz a linha cooc_matrix = np.dot(feature_matrix, feature_matrix.T) é fundamental pois multiplica a matriz pela sua transposta.

O limiar determina quais relações de co-ocorrência são consideradas suficientemente robustas para serem incluídas como arestas no grafo. Genes que coocorrem em pelo menos 5 genomas tendem a representar associações funcionais não casuais [12].

````python
def construir_matriz_coocorrencia(feature_matrix, limiar=1):

    n_genes = feature_matrix.shape[0]
    cooc_matrix = np.dot(feature_matrix, feature_matrix.T)
    cooc_matrix[cooc_matrix < limiar] = 0
    np.fill_diagonal(cooc_matrix, 0) 
    sparse_cooc = csr_matrix(cooc_matrix)
    
    return sparse_cooc
    
limiar_final = 5
adj_matrix = construir_matriz_coocorrencia(feature_matrix, limiar=limiar_final)
n_arestas = adj_matrix.nnz // 2

print(f"Matriz final com limiar = {limiar_final}")
print(f"Nós: {feature_matrix.shape[0]}")
print(f"Arestas: {n_arestas}")
print(f"Densidade: {n_arestas / (feature_matrix.shape[0]**2):.4f}")
````

O resultado indica quantos genomas dois genes compartilham, por exemplo se gene A e B aparecem juntos em 5 genomas, o valor é 5. São em seguida é determinado um linear, tentando respeitar o que é discutido em [11], limiares muito baixos geram grafos excessivamente densos, enquanto limiares muito altos podem fragmentar comunidades funcionais importantes.

## 2. Construção do grafo

### Criando o Grafo NetworkX

A partir grafo no ambiente Network são extraídas as informações topológicas para a GCN. O NetworkX foi selecionado por sua robustez, facilidade de uso e ampla documentação, sendo amplamente utilizado em projetos de bioinformática para manipulação de redes gênicas [11]. Oferece um equilíbrio adequado entre desempenho e facilidade de implementação para grafos de médio porte [9].

Estrutura do grafo:
* Nós: genes
* Arestas: co-ocorrência significativa
* Atributos dos nós: número de genomas onde o gene aparece

Parâmetros essenciais:

* feature_matrix: matriz binária de presença/ausência (shape: n_genes × n_genomas)
* gene_ids: lista com identificadores únicos para cada gene
* limiar: valor mínimo de co-ocorrência para considerar uma aresta significativa

````python
def criar_grafo_networkx(feature_matrix, gene_ids, limiar=2):

    cooc = np.dot(feature_matrix, feature_matrix.T)
    cooc[cooc < limiar] = 0
    np.fill_diagonal(cooc, 0)
    grafo = nx.Graph()
    
    for i, gene_id in enumerate(gene_ids):
        presence = feature_matrix[i]
        n_presence = presence.sum()
        grafo.add_node(gene_id, 
                   presence_count=int(n_presence),
                   presence_pattern=presence.tolist())
    
    for i in range(len(gene_ids)):
        for j in range(i+1, len(gene_ids)):
            if cooc[i, j] > 0:
                grafo.add_edge(gene_ids[i], gene_ids[j], 
                          weight=int(cooc[i, j]))
    
    return grafo

grafo = criar_grafo_networkx(feature_matrix, gene_ids, limiar = limiar_final)

print(f"Nós: {grafo.number_of_nodes()}")
print(f"Arestas: {grafo.number_of_edges()}")
print(f"Componentes conectados: {nx.number_connected_components(grafo)}")

degrees = [d for n, d in grafo.degree()]
print(f"Grau médio: {np.mean(degrees):.2f}")
print(f"Grau máximo: {max(degrees)}")
print(f"Grau mínimo: {min(degrees)}")
````

Os 1361 componentes conectados é um número alto que indica que o grafo é bastante fragmentado. O grau médio de 955.52 em um grafo com 3.141 nós indica que cada gene se conecta, em média, a aproximadamente 30% dos outros genes. Este valor alto confirma a observação anterior sobre a alta densidade do grafo. Grau mínimo sendo zero confirma a existência de nós genes isolados e biologicamente estes podem representar: genes raros ou específicos de linhagem; genes adquiridos horizontalmente e ainda não integrados à rede de co-ocorrência. O cálculo da densidade como n_arestas / (n_nós²), resultando em 1500644 / 9865881 ≈ 0.1521. Este valor é consistente com o esperado para um pangenoma de bactéria de nicho restrito [16].

## Visualização simplificada
Pela grande quantidade de dados a visualização do grafo como um todo é prejudicada, porem ainda é uma possíbilidade uma construção reduzida, o que foi feito na função abaixo.

````python
def visualizar_grafo(grafo, max_nodes=100):

    degrees = dict(grafo.degree())
    top_nodes = sorted(degrees, key=degrees.get, reverse=True)[:max_nodes]
    sub_grafo = grafo.subgraph(top_nodes)
    
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(sub_grafo, k=2, iterations=50)
    
    nx.draw_networkx_nodes(sub_grafo, pos, 
                          node_color='red',
                          node_size=30,
                          alpha=0.7)
    
    nx.draw_networkx_edges(sub_grafo, pos, alpha=0.2, edge_color='gray', width=0.5)
    
    plt.title(f'Grafo de Co-ocorrência de Genes\n(nós: {sub_grafo.number_of_nodes()}, arestas: {sub_grafo.number_of_edges()})')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

visualizar_grafo(grafo)
````

Esta densidade visual é considerada alta para padrões de grafos biológicos [9]. Em comparação, redes de co-expressão gênica típicas apresentam densidades entre 0,01 e 0,05 [11]. A alta densidade observada reforça o que já estava sendo discutido sobre uma estrutura de pangenoma relativamente fechada, característica esperada para Helicobacter pylori, uma bactéria com nicho ecológico restrito [16].

## 3. Conversão para PyTorch Geometric

### Converção do grafo para o Formato PyTorch Geometric (PyG)

A função abaixo representa o ponto de integração entre as estruturas de dados do NetworkX e o PyTorch Geometric. Esta conversão foi feita conforme documentado em [14], o PyG foi especificamente projetado para operações eficientes em grafos, implementando camadas convolucionais otimizadas e suporte nativo a GPUs. A conversão para o formato torch_geometric.data.Data é necessária para aplicar a GCN aos dados de pangenoma [1].


Esta conversão preserva a estrutura de arestas (edge_index), atributos dos nós do tipos int, float e str,e atributos das arestas. A conversão automática nem sempre é eficiente para grafos muito grandes. Conforme [7], para grafos com mais de 10.000 nós, é recomendável construir o objeto Data manualmente para evitar overhead de conversão.

````python
def converter_para_pyg(grafo, feature_matrix, gene_ids):

    data = from_networkx(grafo)
    
    node_order = list(grafo.nodes())
    
    gene_to_idx = {gene_id: i for i, gene_id in enumerate(gene_ids)}
    
    features_reordered = []
    for node in node_order:
        if node in gene_to_idx:
            features_reordered.append(feature_matrix[gene_to_idx[node]])
            
    data.x = torch.tensor(np.array(features_reordered), dtype=torch.float)
    
    presence_counts = np.array([grafo.nodes[node]['presence_count'] for node in node_order])
    
    max_presence = presence_counts.max()
    thresholds = [max_presence * 0.1, max_presence * 0.5]
    
    labels = np.zeros(len(presence_counts), dtype=int)
    labels[presence_counts > thresholds[0]] = 1
    labels[presence_counts > thresholds[1]] = 2
    
    data.y = torch.tensor(labels, dtype=torch.long)
    
    print(f"Nós: {data.num_nodes}")
    print(f"Atributos por nó: {data.num_node_features}")
    print(f"Arestas: {data.num_edges}")
    print(f"Classes: {torch.unique(data.y).tolist()}")
    print(f"Distribuição: {torch.bincount(data.y)}")
    
    return data, node_order

data, node_order = converter_para_pyg(grafo, feature_matrix, gene_ids)
````

A categorização de genes em raros (<10%), moderados (10-50%) e comuns (>50%) é amplamente utilizada em estudos de pangenômica. Esta abordagem permite identificar [12]:

* Genes core (comuns): Geralmente envolvidos em funções essenciais [16]
* Genes acessórios (moderados): Associados a adaptação a nichos específicos
* Genes raros (nuvem): Frequentemente adquiridos por transferência horizontal

Nesse momento a informação de distribuição das classes aponta que não é perfeitamente balanceada (classe 2 com 1374 genes, classe 0 com 1360, classe 1 com apenas 407), porém como essa análise segue fins didáticos, isso não será alterado mas existem ferramentas da bioingormática para contornar esse problema, pois o modelo tende a ignorar classes minoritárias [7]. 

## 4. Divisão treino-validação-teste

### Divisão em treino e teste
 
Diferentemente de abordagens tradicionais de machine learning onde as amostras são independentes e identicamente distribuídas (i.i.d.), em GNNs a divisão deve respeitar a estrutura do grafo [7]. Conforme documentado no PyTorch Geometric [14], a abordagem padrão para classificação de nós é criar máscaras booleanas que indicam quais nós pertencem a cada subconjunto. Estas máscaras são então utilizadas durante o treinamento para calcular a perda (loss) apenas sobre os nós de treino, e durante a avaliação para medir o desempenho nos nós de validação e teste.

````python
def criar_mascaras_treinamento(data, train_ratio=0.7, val_ratio=0.15):

    n_nodes = data.num_nodes
    indices = list(range(n_nodes))
    
    train_idx, temp_idx = train_test_split(indices, test_size=(1 - train_ratio), random_state=SEMENTE_ALEATORIA)
    val_idx, test_idx = train_test_split(temp_idx, test_size=val_ratio/(1-train_ratio), random_state=SEMENTE_ALEATORIA)
    
    train_mask = torch.zeros(n_nodes, dtype=torch.bool)
    val_mask = torch.zeros(n_nodes, dtype=torch.bool)
    test_mask = torch.zeros(n_nodes, dtype=torch.bool)
    
    train_mask[train_idx] = True
    val_mask[val_idx] = True
    test_mask[test_idx] = True
    
    data.train_mask = train_mask
    data.val_mask = val_mask
    data.test_mask = test_mask
    
    print(f"Treino: {train_mask.sum().item()} nós ({train_mask.sum().item()/n_nodes*100:.1f}%)")
    print(f"Validação: {val_mask.sum().item()} nós ({val_mask.sum().item()/n_nodes*100:.1f}%)")
    print(f"Teste: {test_mask.sum().item()} nós ({test_mask.sum().item()/n_nodes*100:.1f}%)")
    
    return data

data = criar_mascaras_treinamento(data)
````

O código primeiro determina o número total de nós (3.141 genes) e cria uma lista com todos os índices. Utilizando a função train_test_split do scikit-learn, ele realiza uma primeira divisão separando 70% dos índices para treinamento e os 30% restantes para um conjunto temporário [13]. Em seguida, o conjunto temporário é novamente dividido, sendo 50% destinado à validação e 50% ao teste, resultando na proporção final 70/15/15. Três tensores booleanos são então criados e preenchidos com True nos índices correspondentes a cada conjunto, formando as máscaras train_mask, val_mask e test_mask. Estas máscaras são atribuídas ao objeto data do PyTorch Geometric [14] e, durante o treinamento, são utilizadas para calcular a perda (loss) considerando apenas os nós de treino, enquanto as máscaras de validação e teste são empregadas para monitorar o desempenho do modelo sem influenciar os gradientes [1].

## 5. Arquitetura da GCN

### Definir o Modelo GCN
A definição do modelo de é responsável por aprender representações vetoriais (embeddings) dos genes que capturem tanto seus padrões individuais de presença/ausência quanto suas relações estruturais de co-ocorrência [1]. Foi construida uma GCN para classificação de genes baseado em seu padrão de coocorrência. A arquitetura utilizada foi baseada no que foi implementado com a biblioteca PyTorch Geometric, amplamente utilizada para aprendizado profundo em grafos [14]. O modelo foi composto por três camadas convolucionais de grafo (Graph Convolution Layers), utilizando a operação GCNConv proposta por Kipf e Welling e implementada no PyTorch Geometric [14]. A arquitetura escolhida segue os princípios fundamentais das GCNs descritos em [1, 7], que destacam a eficácia deste modelo para tarefas de classificação de nós em grafos de médio porte como o deste estudo.

#### Formulação Matemática

A operação convolucional implementada pela classe `GCNConv` do PyTorch Geometric segue a formulação seminal definida pela seguinte equação [1, 14]:

$$X' = \hat{D}^{-1/2} \hat{A} \hat{D}^{-1/2} X W$$

Onde:

- **$X$** : Matriz de características dos nós (features) com dimensões nós × características

- **$\hat{A}$** : Matriz de adjacência com self-loops, calculada como $\hat{A} = A + I_N$  
  - $A$ é a matriz de adjacência original do grafo  
  - $I_N$ é a matriz identidade (adiciona auto-conexões)

- **$\hat{D}$** : Matriz de grau diagonal de $\hat{A}$, onde $\hat{D}_{ii} = \sum_j \hat{A}_{ij}$

- **$W$** : Matriz de pesos treinável da camada convolucional

- **$X'$** : Matriz de características resultante após a convolução

### Interpretação da Equação

A operação realiza essencialmente uma média normalizada das características dos nós vizinhos [1, 7]. O termo $\hat{D}^{-1/2} \hat{A} \hat{D}^{-1/2}$ funciona como um operador de normalização simétrica que preserva a escala das características independentemente do grau do nó, pondera as contribuições dos vizinhos e garante estabilidade numérica durante o treinamento [1, 14].

````python
class GeneGCN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, dropout=0.5):
        super(GeneGCN, self).__init__()
        
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)
        self.conv3 = GCNConv(hidden_dim, output_dim)
        self.dropout = dropout
        
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        # Primeira camada
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        
        # Segunda camada
        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        
        # Camada de saída
        x = self.conv3(x, edge_index)
        
        return F.log_softmax(x, dim=1)

input_dim = data.num_node_features
hidden_dim = 64
output_dim = 3  
dropout = 0.5

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GeneGCN(input_dim, hidden_dim, output_dim, dropout).to(device)
data = data.to(device)

print(f"Input dim: {input_dim}")
print(f"Hidden dim: {hidden_dim}")
print(f"Output dim: {output_dim}")
print(f"Parâmetros totais: {sum(p.numel() for p in model.parameters()):,}")
````

Essa é uma GCN mais simples possível, com três camadas.

* **Camada convolucional:** executa a convolução em cada nó para aprender suas conexões.
* **Camada de ativação não linear:** aplica uma função de ativação como ReLU à saída da convolução.
* **Camada linear de saída:** camada final soma as saídas para gerar uma previsão final.

Primeiro, uma convolução é executada usando cada nó no gráfico da camada de convolução. Essa camada convolucional utiliza informações de funcionalidade dos vizinhos de cada nó e as agrega, atualizando os pesos associados a cada nó. Em seguida, uma função de ativação não linear, a ReLU, é aplicada à saída da camada de convolução. Para obter a melhor precisão, uma rede pode usar múltiplas convoluções e camadas de ativação não lineares empilhadas juntas. Por fim, uma camada linear de saída é usada para prever a qual classe um gráfico provavelmente pertencerá [4].

A escolha de 64 dimensões para o espaço latente e dropout de 0,5 reflete um equilíbrio entre capacidade representacional e regularização. A implementação no PyTorch Geometric [14] demonstra a aplicação prática dos conceitos teóricos de GCNs a um problema real de bioinformática, servindo como base sólida para experimentações futuras com diferentes arquiteturas e hiperparâmetros.

## 6. Treinamento do modelo

### Funções de Treinamento e Avaliação

Diferentemente de redes neurais tradicionais que processam batches de amostras independentes, estas funções operam sobre o grafo completo a cada época, utilizando máscaras para distinguir entre nós de treino, validação e teste [14]. As funções train_epoch, evaluate e treinar_gcn_completo constituem o núcleo do processo de aprendizado da GCN, implementando o ciclo completo de treinamento, validação e teste [1, 7, 14].

#### Função train_epoch o ciclo de treinamento por época
Primeiro o modelo é colocado em modo de treinamento, ativando camadas como dropout que se comportam diferentemente durante o treinamento e inferência [7]. Um aspecto crucial é que, embora a perda seja calculada apenas sobre os nós de treino usando data.train_mask, o forward pass completo model(data) processa todos os nós do grafo [14]. Isto significa que as características de nós de validação e teste influenciam as ativações dos nós de treino através da propagação de mensagens, alé disso informações estruturais de todo o grafo são utilizadas durante o treinamento, é imortante destacar que este é um design intencional das GNNs, não uma limitação [1].

#### Função evaluate para avaliação de desempenho

o modelo passa para o modo de avaliação, desativando camadas específicas de treinamento como o dropout, o que garante que as predições sejam determinísticas e reprodutíveis [7]. É desabilitado o cálculo de gradientes, economizando memória e acelerando o processamento, já que não há retropropagação durante a avaliação [13]. Seguendo é executado o forward pass completo, produzindo log-probabilidades para todos os nós com dimensão [3141, 3]. Em seguida, seleciona a classe de maior probabilidade para cada nó, resultando em um tensor [3141] com valores 0, 1 ou 2 [13], compara as predições com os rótulos verdadeiros apenas para os nós indicados pela máscara, gerando um tensor booleano de acertos/erros. A acurácia é calculada a partir da proporção de predições corretas no conjunto avaliado além da acurácia, também é calculado o F1-score utilizando a função f1_score da biblioteca scikit-learn, que combina precisão e recall em uma única métrica [7], por fim, calcula a perda para o mesmo conjunto [14].

#### Função para treinar GCN

A função do treinamento do modelo, configura o otimizador Adam. Um dicionário "history" é criado para armazenar as métricas de perda, acurácia e F1-score tanto para o conjunto de treino quanto para o de validação ao longo das épocas, permitindo a análise da evolução do aprendizado. O algoritmo implementa early stopping com paciência de 30 épocas, a cada iteração a acurácia de validação é comparada com a melhor já registrada e se houver melhora, o estado atual do modelo é salvo e o contador de paciência é zerado, caso contrário, o contador é incrementado e, ao atingir o limite, o treinamento é interrompido [1,7]. Ao final do treinamento, o melhor estado do modelo (aquele que obteve a maior acurácia de validação) é restaurado e uma avaliação final é realizada no conjunto de teste, calculando-se acurácia, F1-score e perda [1,7].


**Obesrvação sobre o Adam:** O Adam mantém dois momentos para cada parâmetro do modelo, a média dos gradientes (primeiro momento, m) e a média dos quadrados dos gradientes (segundo momento, v), que são atualizados a cada iteração com taxas de decaimento β₁ (padrão 0.9) e β₂ (padrão 0.999) [17]. Essas médias são então corrigidas para compensar o viés inicial e utilizadas para atualizar os parâmetros. Em resumo, o Adam adapta a taxa de aprendizado para cada parâmetro individualmente com base no histórico de gradientes [2], o que o torna eficaz para uma ampla gama de problemas.

````python
def train_epoch(model, data, optimizer):
    model.train()
    optimizer.zero_grad()

    out = model(data)
    
    loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])
    
    loss.backward()
    optimizer.step()
    
    return loss.item()


def evaluate(model, data, mask):
    model.eval()
    with torch.no_grad():
        out = model(data)
        pred = out.argmax(dim=1)
        correct = pred[mask] == data.y[mask]
        acc = correct.sum().item() / mask.sum().item()
        
        loss = F.nll_loss(out[mask], data.y[mask]).item()
        
        f1 = f1_score(data.y[mask].cpu(), pred[mask].cpu(), average='macro')
        
    return acc, loss, f1
    
def treinar_gcn(model, data, epochs=200, lr=0.01, weight_decay=5e-4):
    
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    
    history = {
        'train_loss': [], 'train_acc': [], 'train_f1': [],
        'val_loss': [], 'val_acc': [], 'val_f1': [],
        'test_acc': [], 'test_f1': []
    }
    
    best_val_acc = 0
    best_model_state = None
    patience = 30
    patience_counter = 0
    
    for epoch in range(epochs):
        train_loss = train_epoch(model, data, optimizer)
        train_acc, _, train_f1 = evaluate(model, data, data.train_mask)
        
        val_acc, val_loss, val_f1 = evaluate(model, data, data.val_mask)

        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['train_f1'].append(train_f1)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        history['val_f1'].append(val_f1)
        
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_model_state = model.state_dict().copy()
            patience_counter = 0
        else:
            patience_counter += 1
        
        if (epoch + 1) % 20 == 0:
            print(f"Epoch {epoch+1:3d}/{epochs} | "
                  f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | Train F1: {train_f1:.4f} | "
                  f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | Val F1: {val_f1:.4f}")
        
        if patience_counter >= patience:
            print(f"\nEarly stopping na época {epoch+1}")
            break
    
    model.load_state_dict(best_model_state)
    
    test_acc, test_loss, test_f1 = evaluate(model, data, data.test_mask)
    print(f"Melhor acurácia de validação: {best_val_acc:.4f}")
    print(f"Acurácia no teste: {test_acc:.4f}")
    print(f"F1-score no teste: {test_f1:.4f}")
    print(f"Loss no teste: {test_loss:.4f}")
    
    return history, test_acc, test_f1

history, test_accuracy, test_f1 = treinar_gcn(model, data, epochs=200)
print(f"\nResumo Final:")
print(f"Acurácia no teste: {test_accuracy:.4f}")
print(f"F1-score no teste: {test_f1:.4f}")
````

O modelo atingiu bom desempenho já nas primeiras épocas, com early stopping na época 35, indicando que o modelo converge rapidamente para este dataset. A acurácia foi consistente, com uma pequena diferença entre acurácia de validação (87,47%) e teste (87,71%), sugerindo que não há overfitting significativo [7]. A perda também se mostrou balanceada, com os valores de loss entre treino (0,2986), validação (0,2842) e teste (0,2828) sendo similares, outro indicador de boa generalização. No entanto, uma análise mais aprofundada revela uma diferênça importante entre as métricas, enquanto a acurácia atingiu 87,71% no teste, o F1-score ficou em 0,6255. Esta diferença ocorre porque a acurácia pode ser enganosa quando as classes são desbalanceadas, o modelo que acerta muitas amostras das classes majoritárias (classe 0 com 1360 genes e classe 2 com 1374 genes) e erra todas as amostras da classe minoritária (classe 1 com apenas 407 genes) ainda pode apresentar alta acurácia. O F1-score, ao combinar precisão e recall, penaliza este comportamento, revelando que o modelo teve dificuldade em classificar corretamente os genes da classe intermediária, consistente com a observação anterior de que nenhum gene foi previsto como classe 1. Assim, embora o modelo generalize bem para as classes extremas (genes raros e comuns), ele falha em distinguir a classe moderada, indicando a necessidade de ajustes na arquitetura ou na estrutura do grafo para capturar esta distinção mais sutil.

## Análise de Importância dos Genes

A identificação dos genes que o modelo classifica com maior confiança em cada categoria (raro, moderado, comum), é um tipo de análise que permite validar biologicamente as predições do modelo, verificando se genes conhecidamente conservados ou raros estão sendo adequadamente classificados [9, 11].

````python
def analisar_genes_por_classe(model, data, node_order, gene_ids):
    
    model.eval()
    with torch.no_grad():
        probs = torch.exp(model(data))  
        predictions = probs.argmax(dim=1)
        max_probs = probs.max(dim=1)[0]
    
    predictions_np = predictions.cpu().numpy()
    max_probs_np = max_probs.cpu().numpy()
    
    print("Análise de gene por classe")
    print("")
    
    for classe in range(output_dim):
        mask = predictions_np == classe
        indices = np.where(mask)[0]
        
        nome_classe = {0: "Raros", 1: "Moderado", 2: "Comum"}.get(classe, "_")
        
        print(f"Classe {classe} - {nome_classe}: {mask.sum()} genes")
        print("")
        
        if len(indices) > 0:
            confianca_indices = indices[np.argsort(max_probs_np[indices])[::-1][:5]]
            
            for idx in confianca_indices:
                gene_name = node_order[idx]
                confidence = max_probs_np[idx]
                presence = grafo.nodes[gene_name]['presence_count']
                print(f"{gene_name:<45} | classe predita: {classe} | confiança: {confidence:.3f} | presença: {presence}")
        else:
            print("Nenhum gene classificado nesta categoria")
            print("")
    
    return predictions_np, max_probs_np

predictions_np, max_probs_np = analisar_genes_por_classe(model, data, node_order, gene_ids)
````

````python
def analisar_genes_por_classe(model, data, node_order, gene_ids):

    model.eval()
    with torch.no_grad():
        probs = torch.exp(model(data))  
        predictions = probs.argmax(dim=1)
        max_probs = probs.max(dim=1)[0]
    
    predictions_np = predictions.cpu().numpy()
    max_probs_np = max_probs.cpu().numpy()
    
    for classe in range(output_dim):
        mask = predictions_np == classe
        indices = np.where(mask)[0]
        
        
        if len(indices) > 0:
            confianca_indices = indices[np.argsort(max_probs_np[indices])[::-1][:5]]
            
            for idx in confianca_indices:
                gene_name = node_order[idx]
                confidence = max_probs_np[idx]
                presence = grafo.nodes[gene_name]['presence_count']
                print(f"{gene_name} (confiança: {confidence:.3f}, presença: {presence})")
    
    return predictions_np, max_probs_np 

predictions_np, max_probs_np = analisar_genes_por_classe(model, data, node_order, gene_ids)
````

Essa integração representa uma vantagem importante das GCNs em relação a métodos tradicionais baseados apenas em atributos independentes dos genes [1,7,10].


Para testes futuros pode ser interesante fazer um teste de limiares, de forma a avaliar o desempenho do modelo com limiares progressivamente maiores (10, 15, 20) para identificar o ponto ótimo entre preservação de informação e prevenção de over-smoothing. Além de adotar métricas alternativas de similaridade, implementar o coeficiente de Jaccard ou a correlação de Pearson como medidas de co-ocorrência [11].

## Conclusão

A abordagem de utilizar uma Rede Neural Convolucional de Grafos (GCN) para classificar genes do pangenoma de *Helicobacter pylori* foi bem-sucedida em termos de acurácia, mas revelou limitações importantes relacionadas à estrutura do grafo e à escolha do limiar, indicado pelo F1-Score. O grafo construído a partir da matriz de co-ocorrência com limiar 5 mostrou-se extremamente denso, o que segundo a literatura [7], pode levar ao fenômeno de over-smoothing em GCNs, onde as representações dos nós tornam-se excessivamente similares, prejudicando a capacidade discriminatória do modelo. Isso explica por que o modelo não conseguiu distinguir a classe intermediária.

Essa densidade excessiva provavelmente contribuiu para que a classe intermediária (genes moderados) não fosse identificada pelo modelo, resultando em apenas duas classes na predição final. Ademais, a escolha do limiar de co-ocorrência igual a 2 mostrou-se inadequada para revelar a estrutura comunitária do pangenoma, conforme discutido por Ali et al. (2019), que destacam a importância de uma seleção criteriosa de limiares em redes de co-expressão. Sugere-se, portanto, testar limiares mais elevados (5, 10 ou 20) para testar uma abordadem onde é reduzida a densidade do grafo, e quem sabe melhorar a eficácia da GCN e permitir uma classificação mais refinada que possa distinguir adequadamente as três categorias de genes (raro, moderado, comum), além de fornecer uma visualização mais informativa da topologia da rede gênica.

Todavia o foco do projeto de estudar e explorar a construção em vertente ditática para redes convolucionais de grafos, para te de um grupo maior de Redes Neurais de Grafos foi bem sucedida, sendo possível estruturar um guia geral do que deve ser feito para aplicar esse modelo de maneira eficaz a um problema no qual seja pertinente, além de discutir os principais aspectos que podem levar esse tipo de análise ser relevante entre os inumeros problemas que exigem uma forma gráfica de resolução.
