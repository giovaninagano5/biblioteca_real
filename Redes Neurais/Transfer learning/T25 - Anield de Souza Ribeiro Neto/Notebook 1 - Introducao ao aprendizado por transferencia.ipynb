## 1. A Construção do Conhecimento 

<p style="text-align: justify;"> &emsp; Durante toda a nossa vida estamos em um constante processo de aprendizado. Conscientes ou não, todos os dias vemos novas informações, seja observando novos padrões, ouvindo novas palavras ou aprendendo novos gestos ou técnicas. Mesmo sem notarmos, absorvemos essas informações e passamos a utilizá-las em nosso cotidiano, seja diretamente ou como ponto de partida para o aprendizado de novas informações. Essa é a base de como o nosso conhecimento é construído. </p>

<p style="text-align: justify;"> &emsp; Quando crianças aprendemos a andar de bicicleta, mas quando já adultos vamos aprender a andar de motocicleta, não aprendemos novamente a nos equilibrar, apenas adaptamos e construímos em cima de como aprendemos a nos equilibrar quando criança. Esse, claro, é apenas um exemplo de muitos que podemos imaginar ao realizar esse exercício mental. O que irei fazer ao longo deste trabalho é, também, exemplo desse processo: tentarei utilizar o conhecimento prévio do leitor para simplificar, e acelerar, seu entendimento do que ensinarei. </p>

<p style="text-align: justify;"> &emsp; Em aprendizado de máquina, redes neurais artificiais - que chamarei, por conveniência, apenas como redes neurais - são modelos computacionais com estrutura inspirada no sistema de neurônios e sinapses do cérebro, com o objetivo de construir um modelo matemático se baseando no processo de pensamento e aprendizado humano. Se as redes se inspiram no modelo de pensamento humano, estudaremos neste trabalho, então, como podemos utilizar a ideia de adaptação do conhecimento prévio para a construção de novos saberes para facilitar o treino de redes neurais. Chamaremos esse processo de utilização do conhecimento prévio, no contexto de aprendizado de máquina e redes neurais, de <b>aprendizado por transferência</b>. Iremos começar refletindo por que o aprendizado por trasnferência pode ser vantajoso no campo das redes neurais. </p>


## 2. Os Desafios para o Treino de Modelos de *Deep Learning*

<p style="text-align: justify;"> &emsp; Redes neurais são compostas por uma camada de entrada, que consiste nos dados apresentados à rede uma camada de saída, representando o(s) resultado(s) da rede, e uma ou mais camadas ocultas, nas quais um ou mais neurônios artificiais aplicam pesos, vieses e uma função de ativação aos dados que chegam a eles e passam o resultado para a próxima camada [1] (Figura 1). </p>

       style="width: 500px;">
  <p style="text-align: center;"><em>Figura 1: Forma de uma rede neural de uma camada oculta. Fonte: Cburnett - Obra do próprio., CC BY-SA 3.0 [1].</em></p>
</div>

<p style="text-align: justify;"> &emsp; <i>Deep learning</i> é a área de aprendizado de máquina que utiliza de redes neurais com múltiplas camadas ocultas (Figura 2), chamadas de redes neurais profundas, para resolver problemas como classificação e regressão [2]. O uso de redes neurais profundas permite que diferentes camadas trabalhem diferentes graus de complexidade e abstração dos dados, formando uma hierarquia de níveis que permite que a rede extraia mais informações dos dados.</p>

<p style="text-align: justify;"> &emsp; Redes neurais profundas têm apresentado performance extraordinária em diversos campos. O AlphaFold, desenvolvido pela Google DeepMind, foi capaz de prever com precisão nunca antes obtida a estrutura terciária de sequências de aminoácidos, e o AlphaZero, também da Google DeepMind, foi capaz de superar os melhores jogadores de diversos jogos de tabuleiro como Xadrez e o Go [3,4]. Também foram fundamentais para o desenvolvimento do processamento de linguagem natural e dos grandes modelos de linguagem [2].</p>

</div>

<p style="text-align: justify;"> &emsp; O treino de redes neurais profundas, entretanto, requer um grande montante de dados e custo computacional. Ajustar os parâmetros de uma rede profunda exige tempo, muitos dados registrados e um alto poder de processamento computacional, requisitos que aumentam junto da complexidade da rede. Assim, apesar de suas vantagens, criar esses modelos do zero nem sempre é uma tarefa viável, principalmente em projetos de curto prazo ou em que os dados sejam de difícil obtenção. Nesse contexto, podemos retomar o conceito de aprendizado por transferência. </p>

## 3. O Aprendizado por Transferência

<p style="text-align: justify;"> &emsp; A ideia geral do aprendizado por transferência é bem simples: partimos de um modelo pré treinado e o adaptamos ao caso que estamos estudando. Essa adaptação pode ser feita de duas formas principais, treinando a rede inteira partindo dos pesos já ajustados no treino inicial, processo chamado <i>fine-tuning</i>, ou apenas alterando a camada de saída para ajustar ao novo problema, nesse caso a rede inicial funciona como um extrator de <i>features</i> [6]. Iremos estudar como implementar essas estratégia através do módulo torch no segundo notebook deste trabalho.</p>

<p style="text-align: justify;"> &emsp; Para compreendermos de fato o aprendizado por transferência, é importante que olhemos um pouco para sua definição matemática. Usarei como base a construção apresentada no artigo "Reminder of the First Paper on Transfer Learning in Neural Networks, 1976", Stevo Bozinovski, uma revisão do artigo que introduziu o aprendizado por transferência [7]. Iremos definir o treino de uma rede de classificação simples de modo a alcançar uma relação que mostra o quanto um modelo previamente treinado contribui para um próximo treino.</p>

<p style="text-align: justify;"> &emsp; Considere o conjunto $X=\{x_1,...,x_N\}$ de $N$ padrões, em que um padrão qualquer $x_i \in X$ é um conjunto de valores $\{x_a,...,x_s,...x_Z\}$ dos seus atributos $\{a,...,s,...Z\}$. Esse conjunto $X$ será apresentado a uma rede neural de classificação composta por uma camada de neurônios, cada um correspondente a uma classe $C$. Os neurônios da rede neural realizam, para cada padrão $x$, a operação: $y_i=\sum_s{w_{is}x_s-\theta_i}$, em que $y_i$ é o valor de saída do neurônio $i$, $w_{is}$ o vetor de pesos aplicados sobre a entrada e $\theta_i$ um vetor de vieses. Essa operação pode ser interpretada como a indicação do neurônio da proximidade do conjunto $x$ apresentado à sua classe $C_i$, em que maior sua saída maior a proximidade. A saída dessa rede é dada por um seletor de máximo, em que a classe prevista é dada pela classe do neurônio de maior valor $y$. Ou seja, uma classe $C_i$ é atribuida quando $y_i(x_i)>y_j(x_i)$ em que $j$ é um membro do conjunto dos neurônios tal que $j \neq i$.</p>

<p style="text-align: justify;"> &emsp; Podemos treinar essa rede de modo a ajustar os pesos $w$. Chamaremos o processo de treino da rede, isto é, a apresentação de um conjunto de classe conhecida à rede, de !(x_i). De forma semelhante, chamaremos testar a rede, isto é averiguar qual classe a rede atribui ao padrão, de ?(x_i). Assim, o treinamento da rede visa alcançar o conjunto de valores de pesos em que !(x_i)=?(x_i). Quando essa igualdade não é verdade, a rede ajusta os pesos seguindo: $w_i = w_i + cx_i$, em que $c$ é uma constante de aprendizado de valor usual 1. O resultado da rede treinada será denominado $y_i^0$ e seus pesos $w_i^0$.</p>

<p style="text-align: justify;"> &emsp; Podemos realizar, então uma segunda tarefa de treino partindo da rede anterior. Vamos considerar dois novos padrões $x_i$ e $x_j$ do mesmo espaço de atributos - falarei um pouco sobre espaços diferentes posteriormente - que pertencem às classes $C_i$ e $C_j$ respectivamente. Chamaremos o resultado de um neurônio $i$ na segunda tarefa de treino como $y_i(x_i/L)$, e seu valor é dado por $y_i(x_i/L)=w_i^0x_i - \theta_i^0 + a_{ij}p_i$, em que $a_{ij}$ representa a correlação entre $x_i$ e $x_j$  dada por $x_i{}^Tx_j$ e $p_i$ o número de vezes que os pesos precisarão de serem ajustados para que o modelo aprenda $x_i$. Assim, novamente temos que a classificação correta é dada quando $y_i(x_i/L)>y_j(x_i/L)$. Temos, então, o ajuste de um modelo partindo de um treino anterior já concluído, corrigindo os pesos da mesma forma realizada no primeiro treino, porém começando com o peso $w^0$. Contudo, podemos, ainda, manipular essa expressão para chegar à relação desejada:</p>

$$
\begin{pmatrix}
  a_{ij} & -a_{ji} \\
  -a_{ij} & a_{ji}
\end{pmatrix} 
\begin{pmatrix}
  p_{i} \\
  p_{j}
\end{pmatrix} >
\begin{pmatrix}
  \tau_{ij}(x_i) \\
  \tau_{ji}(x_j) 
\end{pmatrix}, \, \, \, \text{em que}
\begin{pmatrix}
  \tau_{ij}(x_i) \\
  \tau_{ji}(x_j) 
\end{pmatrix} = 
\begin{pmatrix}
  y_i^0(x_i)-y_j^0(x_i) \\
  y_j^0(x_j)-y_i^0(x_j)
\end{pmatrix}
$$

$$
\text{ou ainda,} \, \, \, Ap > \tau
$$

<p style="text-align: justify;"> &emsp; Como os termos da matriz $A$ representam a estrutura dos dados do problema e a matriz de transferência $\tau$ depende de termos do treino já realizado, já conhecemos essas componentes. Assim, essa expressão nos indica o número de ajustes que serão necessários para ajustar a rede a cada novo padrão ensinado tendo como base o treino inicial, isto é, terminar o segundo treino partindo do primeiro. </p>

<p style="text-align: justify;"> &emsp; Note que trabalhamos com uma rede simples de classificação e um único espaço de atributos ao longo da formulação. As ideia apresentadas, porém, também se estendem a redes mais complexas, com mais camadas e funções de ativação não lineares, feitas as devidas adaptações, e diferentes estratégias de ajuste de pesos, uma vez que a expressão final para o número de ajustes necessários para terminar o segundo treino é válida independentemente da forma com que o ajuste é feito. E a matriz de correlação $A$ pode ser construída com dados em espaços diferentes, contanto que estes sejam projetados em um espaço com número de atributos comum aos dois. Independentemente disso, mais importante que a definição matemática é o valor conceitual da relação entre a proximidade dos atributos e o número de novos ajustes ao modelo.</p>

<p style="text-align: justify;"> &emsp; Essa expressão nos permite entender melhor quando é vantajoso a utilização do aprendizado por transferência. Quanto menor a relação entre o espaço de atributos do primeiro e do segundo treino, menor se torna a contribuição - podendo essa ser negativa, implicando que ela apenas dificultou mais o processo de treino - do primeiro treino para a o segundo, e maior o número de ajustes que serão necessários no modelo. Dessa forma, utilizar o aprendizado por transferência pode ser muito positivo quando partimos de um modelo treinado em problema semelhante ao que estamos trabalhando: um modelo treinado para identificar facas e garfos pode aprender com mais facilidade a identificar colheres. Mas quando afastamos os domínios das tarefas, se torna menos vantajoso a utilização do modelo base, valendo a busca por outro modelo mais adequado. </p>

## 4. Conclusão

<p style="text-align: justify;"> &emsp; Introduzidos ao aprendizados por transferência, podemos enfim responder a questão mais importante: quando utilizar o aprendizado por transferência? Em geral, essa estratégia é muito útil em casos de escassez de dados, necessidade de redes muito complexas, que seriam inviáveis de treinar nas condições do projeto, e quando já existem modelos muito bem treinados em problemas próximos ao nosso. 

<p style="text-align: justify;"> &emsp; Com isso, construímos a base do que é o aprendizado por transferência e sua relevância no treino de modelos complexos.
